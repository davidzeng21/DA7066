```python
import os
import pandas as pd
data_file = os.path.join('data', 'data.csv')
results_df = pd.read_csv(data_file)
xdata = results_df['x'] # extracting the x values
ydata = results_df['val'] # extracting the measurement values
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(xdata, ydata)
print("Slope = ", slope)
print("Intercept = ", intercept)
print("R-squared = ", r_value**2)
print("P value = ", p_value)
print("Standard error = ", std_err)
```

```python
pyplot.scatter(xdata, ydata) # xdata and ydata are values from above
xvals = range(-1,2)
yvals = slope*range(-1,2)+intercept
pyplot.plot(xvals, yvals, color='g') # adding regression line
```
